# Model Evaluation and Refinement

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    
    ax1 = sns.kdeplot(RedFunction, color="r", label=RedName)
    ax2 = sns.kdeplot(BlueFunction, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.xlabel('Price')
    plt.ylabel('Proportion of Houses')
    plt.show()
    plt.close()
```

<h2 id="ref1">Training and Testing</h2>

```python
x_data=df.drop('price',axis=1)
y_data = df['price']
```

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.20, random_state=0)


print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn.linear_model import LinearRegression
lre=LinearRegression()
lre.fit(x_train[['area']], y_train)
```

```python
lre.score(x_test[['area']], y_test)
```

```python
lre.score(x_train[['area']], y_train)
```

<h2>Cross-Validation Score</h2>

```python
from sklearn.model_selection import cross_val_score
Rcross = cross_val_score(lre, x_data[['area']], y_data, cv=4)
print("The mean of the folds are", Rcross.mean(), "and the standard deviation is" , Rcross.std())
```

```python
from sklearn.model_selection import cross_val_predict
yhat = cross_val_predict(lre,x_data[['area']], y_data,cv=4)
yhat[0:5]
```

## Measures for Model Evaluation

<b>R-squared</b>

```python
lre.fit(x_train[['area']], y_train)
print('The R-square for training data is: ', lre.score(x_train[["area"]], y_train))
```

```python
lre.fit(x_test[['area']], y_test)
print('The R-square for test data is: ', lre.score(x_test[["area"]], y_test))
```

<b>Mean_Squared_Error</b>

```python
from sklearn.metrics import mean_squared_error
mse_train = mean_squared_error(y_train, yhat_train)
print('The mean square error of price and predicted value of training set is: ', mse_train)

mse_test = mean_squared_error(y_test, yhat_test)
print('The mean square error of price and predicted value of test set is: ', mse_test)
```
